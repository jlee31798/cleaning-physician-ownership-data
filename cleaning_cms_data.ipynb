# Name Suffix

```python
df['name_suffix'].unique()
```

Alright. This can get complicated. Not only is there inconsistent capitlizations, some suffixes have dots and some don't... Let me do some research really quick.

So after some googling, here are some findings:
- 'Jr' and 'Sr' should have a period at the end when displaying in a full name. e.g. "John F. Kennedy Jr."
- MD, OD, DPM, DMD all need periods in between. e.g. 'M.D.'
- 'Mr' needs a period


```python
import fuzzywuzzy
from fuzzywuzzy import process
```

```python
suffixes = df['name_suffix'].unique()
suffixes
```

```python
def match_strings(data, column, string_to_match, min_ratio):
    # get the list of unique strings under the column
    strings = df[column].unique()
    
    # get the top 10 closets matches for the stringt to match
    matches = fuzzywuzzy.process.extract(string_to_match, strings, limit = 10, scorer = fuzzywuzzy.fuzz.token_sort_ratio)
    
    # get the matches that satisfy the min_ratio
    close_matches = [matches[0] for matches in matches if matches[1] == min_ratio]
    
    # get the rows with those close matches
    rows_with_match = df[column].isin(close_matches)
    rows_with_match
    # replace rows with string to match
    df.loc[rows_with_match, column] = string_to_match
    
    print('Finish!')
    
```

```python
# Change all jr's to 'Jr.'
match_strings(df, 'name_suffix', 'Jr.', 100)

#fuzzywuzzy.process.extract('Jr.', suffixes, limit = 10, scorer = fuzzywuzzy.fuzz.token_sort_ratio) 
match_strings(df, 'name_suffix', 'Sr.', 100)

# do this with mr and mrs as well
match_strings(df, 'name_suffix', 'Mr.', 100)

match_strings(df, 'name_suffix', 'Mrs.', 100)


```

```python
# replace 'MD' with 'M.D.'
df.loc[df.name_suffix == 'MD', 'name_suffix'] = 'M.D.'
# replace 'JR MD' with 'Jr., M.D.'
df.loc[df.name_suffix == 'JR MD', 'name_suffix'] = 'Jr., M.D.'
# replace 'DPM' with 'D.P.M.','DMD' with 'D.M.D.' & 'DDS' with 'D.D.S.'
df.loc[df.name_suffix == 'DPM', 'name_suffix'] = 'D.P.M.'
df.loc[df.name_suffix == 'DMD', 'name_suffix'] = 'D.M.D.'
df.loc[df.name_suffix == 'DDS', 'name_suffix'] = 'D.D.S.'
df.loc[df.name_suffix == 'OD', 'name_suffix'] = 'O.D.'

# do this with the number suffixes combined with MD
df.loc[df.name_suffix == 'IIIMD', 'name_suffix'] = 'III M.D.'
df.loc[df.name_suffix == 'II MD', 'name_suffix'] = 'II M.D.'
df.loc[df.name_suffix == 'II MD', 'name_suffix'] = 'II M.D.'

```

# Null Values 

Let's look with the null values!

```python
df.isnull().sum()
```

columns with null values:
    - middle_name
        - Null value here means no middle name which is fine, replace with empty string ('')
    - name_suffix
        - null value means no suffix, replace with empty string('')
    - address_line2
        - null value means the address provided doesn't have a second line of street address, replace with empty string ('')
    - province
        - no province provided, would just have to remove column
    - postal_code
        - with a bit of Google Search, Turns out postal code and Zip Code are essentially the same thing except Postal Code is more frequently used in the UK, and Zip Code is more frequently used here. Remove column
    - specialty
        - This just means physicians don't have a specialty, replace with ('N/A')
    - org_payment_state
        - every org would need a state, might need further investigation
    

```python
df = df.drop('province', axis = 1).drop('postal_code', axis = 1)
df.columns
```

```python
# replace middle_name with ('')
df['middle_name'] = df['middle_name'].fillna('')
# replace name_suffix with ('')
df['name_suffix'] = df['name_suffix'].fillna('')
# replace address_line2 with ('')
df['address_line2'] = df['address_line2'].fillna('')
# replace specialty with ('N/A')
df['specialty'] = df['specialty'].fillna('N/A')
```

```python
# org_payment_state
df[df['org_payment_state'].isnull()]
```

Those without states are organizations that are not in the US. Replace those without states with ('N/A')

```python
df['org_payment_state'] = df['org_payment_state'].fillna('N/A')
```
